```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Enter the length of the box and the quantum number n

```python
box_length = 1
n = 5
```

This is a bunch of code you do not need to change. It includes function definitions etc.

```python
L = box_length**-10
x,y = np.linspace(0, L, 200), np.linspace(0, L, 200) #generates evenly spaced data points
x1,y1 = np.linspace(0, L, 200), np.linspace(0, L, 200) #generates evenly spaced data points
x2,y2 = np.linspace(0, L, 200), np.linspace(0, L, 200) #generates evenly spaced data points
A = (2/L)**0.5 #normalization constant

def psi_2d(a): #computes the wave function (2D)
    psi_2d = A*np.sin((n * np.pi * a)/L)
    return psi_2d 

def psi_2d_sq(a): #computes the probability density (2D)
    psi_2d_sq = (A*np.sin((n * np.pi * a)/L))**2
    return psi_2d_sq

def psi_3d(a,b): #computes the wave function (3D)
    psi_3d = A*np.sin((n * np.pi * a)/L)*np.sin((n * np.pi * b)/L)
    return psi_3d

def psi_3d_sq(a,b): #computes the probability density (3D)
    psi_3d_sq = (A*np.sin((n * np.pi * a)/L)*np.sin((n * np.pi * b)/L))**2
    return psi_3d_sq
```

```python
PSI = psi_2d(x)
PSI_SQ = psi_2d_sq(x)
```

```python
X1,Y1 = np.meshgrid(x1,y1)
psi_3d = np.array([psi_3d(x,y) for x,y in zip(np.ravel(X1), np.ravel(Y1))])

X2,Y2 = np.meshgrid(x2,y2)
psi_3d_sq = np.array([psi_3d_sq(x,y) for x,y in zip(np.ravel(X1), np.ravel(Y1))])
```

```python
PSI_3D =  psi_3d.reshape(X1.shape)
PSI_3D_SQ = psi_3d_sq.reshape(X2.shape)
```

Now here come the plots. Feel free to play around with different settings such as size, color, etc :)

```python
#plotting 1D wave function
plt.figure(figsize=(8,6))
plt.plot(x,PSI,'darkblue')
plt.plot(x,PSI_SQ,'lime')
plt.xlabel('x',fontsize=13)
plt.ylabel('y',fontsize=13)
plt.legend(['$\Psi$','$\Psi^{2}$'],fontsize=13)
plt.title('Particle in a 1D Box',fontsize=13);
```

```python
#plotting the 3D wave function
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X1,Y1,PSI_3D, cmap = 'winter')
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Z')
plt.title('Wavefunction for a Square Box', color='magenta')
```

```python
#plotting the 3D probability density
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X2,Y2,PSI_3D_SQ, cmap = 'winter')
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Z')
plt.title('Probability Density for a Square Box', color='magenta')
```
